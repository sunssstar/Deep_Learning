```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.array([2,4,6,8])
y = np.array([81,93,91,97])
```

```python
plt.scatter(x,y)
plt.show()
```

```python
# 기울기 w 절편 b

# 초기화
w = 0
b = 0

# 학습률(learning rate)
# 임의로 0.03 의미 없음
lr = 0.03

# 반복 횟수
epochs = 2001
```

```python
x = np.array([2, 4, 6, 8])
n = len(x)
print(n)
```

```python
# 경사하강법(sgd)
n = len(x)

# 경사하강법 : 편미분을 하여 오차가 가장 적은 경사(기울기, 가중치)찾는 법

# 경사하강법 반복문
for i in range(epochs):
  # epochs 수 만큼 반복
  y_pred = w*x + b
  error = y - y_pred

  # w = 가중치,
  # b = 편향(bias)
  w_diff = (2/n) * sum(-x*(error)) # 오차(비용)함수를 w로 편미분 한 값
  b_diff = (2/n) * sum(-error)     # 오차(비용)함수를 b로 편미분 한 값

  w = w - (lr*w_diff)                # w의 편미분 값을 학습률(lr)을 곱해서 기존의 w를 업데이트
  b = b - (lr*b_diff)                # b의 편미분 값을 학습률(lr)을 곱해서 기존의 b를 업데이트

  if i %  100 == 0:
    print("epochs = %.f, 기울기 = %.4f, 절편=%.4f"%(i,w,b))
```

```python
# 결과 해석
# w(기울기) : 2.3
# b = 79
# y_pred = 2.3 x(공부시간) + 79
```

```python
y_pred = w*x + b

plt.scatter(x,y)
plt.plot(x,y_pred,c='r')
plt.show()
```

다중선형회귀분석

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x1 = np.array([2,4,6,8])
x2 = np.array([0,4,2,3])
y = np.array([81,93,91,97])
```

```python
# 데이터 분포 확인
fig = plt.figure()
ax = fig.add_subplot(111,projection = '3d')
ax.scatter3D(x1,x2,y)
plt.show()
```

```python
# 변수 설정

# 초기화
w1 = 0
w2 = 0
b = 0

# 학습률(learning rate)
lr = 0.01

# 반복 횟수
epochs = 2001
```

```python
# 경사하강법

n = len(x1)

for i in range(epochs):
  y_pred = w1*x1 + w2*x2 + b
  error = y-y_pred

  # 편미분
  w1_diff = (2/n)*sum(-x1*(error))
  w2_diff = (2/n)*sum(-x2*(error))
  b_diff = (2/n)*sum(-error)

  # 학습률
  w1 = w1 - lr*w1_diff
  w2 = w2 - lr*w2_diff
  b = b - lr*b_diff

  if i % 100 == 0:
    print("epochs:%.f, 기울기 w1:%.4f, 기울기w2:%.4f, 절편b:%.4f"%(i,w1,w2,b))
```

```python
# 실제 점수와 예측 점수 출력

print('실제점수:',y)
print('예측점수:',y_pred)
```

```python
# 텐서플로우 활용, 선형회귀모델

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
x = np.array([2,4,6,8]) # 공부 시간
y = np.array([81,93,91,97 ]) # 시험 점수
```

```python
model = Sequential()

# 출력값, 입력 변수, 분석방법(활성화 함수)
model.add(Dense(1,input_dim = 1, activation = 'linear'))

# 오차 수정을 위해 경사하강법(sgd) 사용
# 오차의 정도를 판단하는 평가방법 --> mse(평균제곱오차)
model.compile(optimizer = 'sgd', loss='mse')

model.fit(x,y,epochs = 2001)
```

```python
# 모델 테스트

hour = 7
prediction = model.predict([hour])

print("%.f 시간을 공부하시면, 예상점수는 %.02f 점입니다."%(hour,prediction))
```

```python
# 다중회귀분석

x = np.array([[2,0],[4,4],[6,2],[8,3]])
y = np.array([81,93,91,97])
```

```python
model = Sequential()

# 출력값, 입력 변수, 분석방법(활성화 함수)
model.add(Dense(1,input_dim = 2, activation = 'linear'))

# 오차 수정을 위해 경사하강법(sgd) 사용
# 오차의 정도를 판단하는 평가방법 --> mse(평균제곱오차)
model.compile(optimizer = 'sgd', loss='mse')

model.fit(x,y,epochs = 2001)
```

```python
# 모델 테스트

hour = 7
private_class = 4
prediction = model.predict([[hour,private_class]])

print("%.f 시간을 공부하시고,%.f시간 과외를 하면 예상점수는 %.02f 점입니다."%(hour,private_class,prediction))
```
