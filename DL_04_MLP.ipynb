```python
# MLP(multi - layer Perceptron : 다층 퍼셉트론)

# 미션 : 다층 퍼센트론으로 XOR 문제 해결

import numpy as np
import pandas as pd
```

```python
# 가중치와 바이어스 설정(임의 값)

w11 = np.array([-2,-2])
w12 = np.array([2,2])
w2 = np.array([1,1])

b1 =3
b2 = -1
b3 = -1
```

```python
# perceptorn

def MLP(x,w,b):
  y = np.sum(w*x)+b
  if y <=0:
    return 0
  else:
    return 1
```

```python
# XOR 문제 해결 위한 게이트 설정

# ANN(NAND,OR) >>XOR

# NAND
def NAND(x1,x2):
  return MLP(np.array([x1,x2]),w11,b1)

# OR
def OR(x1,x2):
  return MLP(np.array([x1,x2]),w12,b2)

# AND
def AND(x1,x2):
  return MLP(np.array([x1,x2]),w2,b3)

# XOR
def XOR(x1,x2):
  return AND(NAND(x1,x2),OR(x1,x2))
```

```python
# x1,x2 값을 번갈아 대입 --> 최종 값 출력

for x in [(0,0),(1,0),(0,1),(1,1)]:
  y = XOR(x[0],x[1])
  print("입력 값 : "+str(x)+" 출력 값 : "+str(y))
```
