```python
# 회귀분석의 기초_ 가장 훌륭한 예측선 그리기

# 최소제곱

import numpy as np
import pandas as pd

# 공부한 시간과 점수 각각 x,y 변수에 넣기

x = np.array([2,4,6,8]) # 공부 시간
y = np.array([81,93,91,97]) # 시험 점수
```

```python
# x,y의 평균값 구하기

mx = np.mean(x)
my = np.mean(y)

print('x 평균값:', mx)
print('y 평균값:', my)
```

```python
[(i-mx) **2 for i in x]
```

```python
sum([(i-mx) **2 for i in x])
```

```python
# 기울기 공식: y = wx + b
# w(weight) 가중치 b(bias) 편차

# 최소제곱법 분모 만들기

divisor = sum([(i-mx) **2 for i in x])

# 최소제곱법 분자 만들기
def top(x, mx, y, my):
    d = 0
    for i in range(len(x)):
        d += (x[i] - mx) * (y[i] - my)
    return d
dividend = top(x, mx, y, my)

print('분모: ', divisor)
print('분자: ', dividend)
```

```python
# 평균 제곱 오차 구하기

w = dividend/divisor

# 절편 b
b = my -(mx*w)

print('기울기 w', w)
print('y절편 b', b)
print('가장 훌륭한 예측선을 %.2fx+%d 이다'%(w, b))
```

```python
# 평균제곱오차

# 임의의 가상의 기울기 값 w, 절편 b 입력

w = 3
b = 76


x = np.array([2,4,6,8]) #공부한 시간
y = np.array([81,93,91,97]) # 시험 본 점수
```

```python
# 평균제곱오차

# 임의의 가상의 기울기 값 w, 절편 b 입력

#예측함수

def predict(x):
    return w * x + b

# 예측값이 들어갈 빈 리스트 생성
result = []

for i in range(len(x)):
    result.append(predict(x[i]))
    print("공부한 시간: %.2f, 실제 점수: %.2f, 예측 점수: %.2f"%(x[i], y[i], predict(x[i])))
```

```python
# 평균 제곱 오차 함수 이용, 각 y 값 대입 -> 최종값 구하기

n = len(x)

def mse(y,y_pred):
  return (1/n) * sum((y-y_pred)**2)

# 평균제곱 오차
print("평균제곱오차:" +str(mse(y,result)))
```
