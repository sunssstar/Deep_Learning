```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
x = np.array([2,4,6,8,10,12,14])
y = np.array([0,0,0,1,1,1,1])
```

```python
# Define the model
model = Sequential()
model.add(Dense(1, input_dim=1, activation='sigmoid'))  # 1 unit, input shape (1,)

# Compile the model
model.compile(optimizer='sgd', loss='binary_crossentropy')

# Fit the model
model.fit(x, y, epochs=5000)
```

```python
# 모델 테스트 데이터 시각화

plt.scatter(x,y)
plt.plot(x,model.predict(x),'r')
plt.show()
```

```python
# 임의의 학습시간 입력, 합격예상 확률 출력

hour = 7

prediction = model.predict([[hour]])
print('%.d 시간을 공부하면, 합격 예상확률을 %.2f%%입니다.'%(hour,prediction * 100))
```
